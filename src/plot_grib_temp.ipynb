```python
%matplotlib inline
from pyspark import SparkContext
import pygrib
import numpy.ma as ma
import numpy as np
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
```

```python
def loadEOF(fname, numeofs, numobs):
    eofs = np.zeros((numeofs, numobs))
    with open(fname, 'r') as fin:
        fin.readline() # skip the two header lines
        fin.readline()
        for line in fin:
            (i, j, v) = line.split()
            eofs[int(j)-1, int(i)-1] = float(v)
    return eofs

def loadEvals(fname, numeofs):
    evals = np.zeros((numeofs,))
    with open(fname, 'r') as fin:
        fin.readline()
        fin.readline()
        for line in fin:
            (i,j,v) = line.split()
            evals[int(j)-1] = float(v)
    return evals

def extractSST(eof):
    eofpadded = np.zeros((dims,))
    eofpadded[mask == False] = eof
    return eof[0:dimsperlevel]
    
def scaleSST(sst):
    return (sst - sst[SSTmask==False].mean())/sst[SSTmask==False].std()

def visualSST(sst):
    return ma.masked_array(sst.reshape((360,720)), SSTmask.reshape((360,720)))
```

```python
grbs = pygrib.open("base.grb2")
sstoffset = 207
temps = grbs[sstoffset].values
sstlats, sstlons = grbs[sstoffset].latlons()

gribindices = range(1,41)
gribindices.append(207)
gribindices = list(reversed(gribindices))

dimsperlevel = 360*720
dims = dimsperlevel * 41
mask = np.zeros((dims,)) < 0

for index in range(41):
    mask[index*dimsperlevel:(index+1)*dimsperlevel] = grbs[gribindices[index]].data()[0].mask.reshape((dimsperlevel,))
    
SSTmask = mask[0:dimsperlevel]

eofs = loadEOF("/home/ubuntu/large-scale-climate/data/colEOFs", 20, 6349676)
evals = loadEvals("/home/ubuntu/large-scale-climate/data/evalsEOFs", 20)
```

```python
fig = plt.figure()
plt.plot(range(1, 21), evals)
plt.ylabel("Eigenvalues")
plt.xlabel("EOF index")
plt.show()
```

```python
sst = extractSST(eofs[19,:])
visualSST(sst)

fig = plt.figure()
ax = fig.add_axes([0.05,0.05,0.9,0.9])
# create Basemap instance.
# coastlines not used, so resolution set to None to skip
# continent processing (this speeds things up a bit)
m = Basemap(projection='kav7',lon_0=0,resolution=None)
# draw line around map projection limb.
# color background of map projection region.
# missing values over land will show up this color.
m.drawmapboundary(fill_color='0.3')
im1 = m.pcolormesh(sstlons,sstlats,visualSST(scaleSST(sst)),shading='flat',cmap=plt.cm.jet,latlon=True)
# draw parallels and meridians, but don't bother labelling them.
m.drawparallels(np.arange(-90.,99.,30.))
m.drawmeridians(np.arange(-180.,180.,60.))
# add colorbar
cb = m.colorbar(im1,"bottom", size="5%", pad="2%")
# add a title.
ax.set_title('SST portion of first EOF')
plt.show()
```
